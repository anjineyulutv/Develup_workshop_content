<h4>Let's normalize the data and apply linear regression</h4>

```python
cols=['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Avg. Area Number of Bedrooms', 'Area Population', 'Price']
for i in cols:
    df[i]=(df[i]-df[i].min())/(df[i].max()-df[i].min())
```

```python
df.describe()
#0-1
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Avg. Area Number of Bedrooms', 'Area Population']], df['Price'], test_size=0.33, random_state=42)
```

```python
reg = LinearRegression().fit(X_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```
